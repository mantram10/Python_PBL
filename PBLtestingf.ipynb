```python
import numpy as np

def log_transform(X):
    return np.log1p(X)
```

```python
import pickle

with open("final_model.pkl", "rb") as f:
    model = pickle.load(f)
```

```python
print("Enter user details:\n")

data = {
    "Administrative": int(input("Administrative pages: ")),
    "Administrative_Duration": float(input("Administrative duration: ")),
    "Informational": int(input("Informational pages: ")),
    "Informational_Duration": float(input("Informational duration: ")),
    "ProductRelated": int(input("Product related pages: ")),
    "ProductRelated_Duration": float(input("Product related duration: ")),
    "BounceRates": float(input("Bounce rate (0-1): ")),
    "ExitRates": float(input("Exit rate (0-1): ")),
    "PageValues": float(input("Page values: ")),
    "SpecialDay": float(input("Special day (0-1): ")),
    "Month": input("Month (Jan, Feb, Mar...): "),
    "OperatingSystems": int(input("Operating system (1-8): ")),
    "Browser": int(input("Browser (1-13): ")),
    "Region": int(input("Region (1-9): ")),
    "TrafficType": int(input("Traffic type: ")),
    "VisitorType": input("Visitor type (Returning_Visitor/New_Visitor/Other): "),
    "Weekend": input("Weekend (True/False): ") == "True"
}
```

```python
import pandas as pd
user_data = pd.DataFrame([data])
```

```python
prob = model.predict_proba(user_data)[:, 1]

threshold = 0.3
prediction = (prob >= threshold).astype(int)
```

```python
print("\n--- RESULT ---")

print(f"Probability of purchase: {prob[0]:.2f}")

if prediction[0] == 1:
    print("Prediction: User is likely to PURCHASE")
else:
    print("Prediction: User is NOT likely to purchase")
```

```python
prob_value = prob[0]

bar_length = int(prob_value * 50)

print("\nConfidence:")
print("[" + "#" * bar_length + "-" * (50 - bar_length) + "]")
print(f"{prob_value*100:.2f}% chance of purchase")
```

```python
print("\n--- Key Input Summary ---")

print(f"Product Pages: {data['ProductRelated']}")
print(f"Time on Product Pages: {data['ProductRelated_Duration']}")
print(f"Bounce Rate: {data['BounceRates']}")
print(f"Exit Rate: {data['ExitRates']}")
print(f"Page Value: {data['PageValues']}")
```

```python
if data["PageValues"] > 10:
    print("High page value → strong buying signal")

if data["BounceRates"] > 0.1:
    print("High bounce rate → weak engagement")

if data["ProductRelated"] > 20:
    print("Many product pages → interested user")
```

```python
if prob_value > 0.7:
    print("Model is VERY confident")
elif prob_value > 0.4:
    print("Model is moderately confident")
else:
    print("Model has low confidence")
```

```python
print("\n--- Threshold Analysis ---")

for t in [0.5, 0.4, 0.3]:
    pred = (prob >= t).astype(int)
    print(f"Threshold {t}: Prediction = {pred[0]}")
```
